```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/pattaradanai-lks/EPL2021-2022/main/2021-2022.csv"
epl = pd.read_csv(url)
epl
```

# 1. Simple stats 3 questions

```python
#1.1 มีทีมอะไรบ้างใน EPL 2021-2022

pd.DataFrame(epl["HomeTeam"].unique(), columns=['Team'])
```

```python
#1.2 ทีมใดทำประตูในฐานะการเป็นทีมเยือน(Away Team)ต่อ1นัดเยอะที่สุด

team_with_max_home_goals = epl.loc[epl['FTAG'].idxmax()]['AwayTeam']
max_home_goals = epl['FTAG'].max()

print(f"ทีมที่ทำประตูเยอะที่สุดเป็นทีม {team_with_max_home_goals} ด้วยจำนวนประตูทั้งหมด {max_home_goals} ประตู")
```

```python
#1.3 แต่ละทีมทำประตูในฐานะการเป็นทีมเหย้า(HomeTeam)ตลอดทั้งฤดูกาลเป็นจำนวนเท่าไร

total_home_goals_by_team = epl.groupby('HomeTeam')['FTHG'].sum()
total_home_goals_by_team_sorted = total_home_goals_by_team.sort_values(ascending=False)

print(total_home_goals_by_team_sorted)
```

# 2. Advanced 2 questions


```python
#2.1ทีมใดที่โดนนำในครึ่งแรกแล้วสุดท้ายกลับมาชนะได้มากที่สุด

# กรองข้อมูลทีมที่แพ้ในครึ่งแรกแล้วชนะท้ายที่สุด
comeback_teams = epl[(epl['HTR'] == 'A') & (epl['FTR'] == 'H') | (epl['HTR'] == 'H') & (epl['FTR'] == 'A')]

# นับจำนวนการเปลี่ยนแปลงผลลัพธ์ของการแข่งขัน
comeback_counts_by_team = comeback_teams.groupby('HomeTeam').size() + comeback_teams.groupby('AwayTeam').size()

# หาทีมที่มีการcomebacksที่สุด
team_with_max_comebacks = comeback_counts_by_team.idxmax()
max_comebacks = comeback_counts_by_team.max()

print(f"ทีมที่แพ้ในครึ่งแรกแล้วกลับมาชนะมากที่สุดคือทีม {team_with_max_comebacks} ทั้งหมด {max_comebacks} ครั้ง")
```

```python
#2.2ทีมที่ได้โอกาศยิงเยอะที่สุด กับทีมที่ยิงตรงกรอบมากที่สุด และทีมที่ยิงประตูมากที่สุดใช่ทีมเดียวกันหรือไม่ ถ้าไม่ใช่แล้วทีมนั้นคือทีมอะไรบ้าง

total_shots_by_team = epl.groupby('HomeTeam')['HS'].sum() + epl.groupby('AwayTeam')['AS'].sum()
team_with_max_total_shots = total_shots_by_team.idxmax()
max_total_shots = total_shots_by_team.max()

print(f"ทีมที่ได้โอกาสยิง(HS+AS)มากที่สุดคือทีม {team_with_max_total_shots} ด้วยโอกาสการยิงรวม {max_total_shots} ครั้ง")

total_shots_by_team = epl.groupby('HomeTeam')['HST'].sum() + epl.groupby('AwayTeam')['AST'].sum()
team_with_max_total_shots = total_shots_by_team.idxmax()
max_total_shots = total_shots_by_team.max()

print(f"ทีมที่ยิงตรงกรอบ(HST+AST)มากที่สุดคือทีม {team_with_max_total_shots} ด้วยจำนวนการยิงตรงกรอบ {max_total_shots} ครั้ง")

total_shots_by_team = epl.groupby('HomeTeam')['FTHG'].sum() + epl.groupby('AwayTeam')['FTAG'].sum()
team_with_max_total_shots = total_shots_by_team.idxmax()
max_total_shots = total_shots_by_team.max()

print(f"ทีมที่มีจำนวนการยิงประตูรวม(FTHG+FTAG)มากที่สุดคือทีม {team_with_max_total_shots} ด้วยจำนวนการยิงรวม {max_total_shots} ประตู")
```
